```python
def extract_keys(js, field, encode=False):
    try:
        d = json.loads(js)
    except:
        d = {}
    if not encode:
        return d.get(field, None)
    else:
        return json.dumps(d.get(field, ""))
    
def extract_data_dicts( js, first, secund ):
    try:
        d = json.loads(js)
    except:
        d = {}
    #d = json.loads(js)
    try:
        e = d.get(first, {}).get(secund)
    except:
        e = {}
    return e

```

```python
#Information to deliver
#Taken from https://www.jeevangupta.com/python-code-to-get-clickup-space-workspace-and-task-details-using-clickup-api/
    
def get_tasks_details(clickup_api, list_id):
    try:
        url = "https://api.clickup.com/api/v2/list/%s/task?archived=false"%list_id
        #url = "https://api.clickup.com/api/v2/list/{list_id}/task"

        headers = {"Authorization": clickup_api}
        r = requests.get(url = url, headers = headers)
        response_dict = json.loads(r.text)
        #print(response_dict)
 
        tasks = response_dict["tasks"]
 
        clickup_tasks = pd.DataFrame()
        task_list = []
        for task in tasks:
            tmp_dict = {}
            task_id = task['id']
            task_name = task['name']
            task_content = task['text_content']
            #link_id = task['link_id']
 
            tmp_dict["task_name"] = task_name
            tmp_dict["task_id"] = task_id
            tmp_dict["task_content"] = task_content
            #tmp_dict["link_id"] = task['link_id']
            tmp_dict["date_created"] = task['date_created']
            tmp_dict["status"] = task['status']['status']
            tmp_dict["creator"] = task['creator']
            tmp_dict["assignees"] = task['assignees']
            tmp_dict["watchers"] = task['watchers']
 
            #print("\n task name : %s | task id : %s "% (task_name, task_id))
 
            task_list.append(tmp_dict)
 
        clickup_tasks = pd.DataFrame.from_records(task_list)
 
        return clickup_tasks
 
    except:
        print("\n get_tasks_details Failed : ",sys.exc_info())
 

if __name__ == '__main__':
    try:
        print(" ClickUp task list and details process starts")
 
        clickup_api = "pk_3041236_DXOBGO4W2M220TSFKLKVFUDZ50WP4VG9"
        list_id = "900100953154"
 
        Data_CLIENTES = get_tasks_details(clickup_api, list_id)
        #print(Data_CLIENTES)
 
 
        print("\n ClickUp task list and details process Finished.")
    except:
        print("\n ClickUp task list and details process Failed : ", sys.exc_info())
```

        <strong>TRANSFORM:</strong><br>
En la entidad Proyecto vamos a realizar algunas transformaciones: <br>
Campo PRODUCTO: Producto WEB, SEO, RRSS, etc.<br>
● 0 = KD-WEB<br>

```python
def add_product(df):
    list_r = []
    for row in df['task_name']:
        if "KD-WEB" in row:    list_r.append("0")
        elif "KD-SEO" in row:   list_r.append("1")
        elif "KD-ANALITICA" in row:  list_r.append("2")
        elif "KD-RRSS" in row:  list_r.append("3")
        elif "KD-ECOMMERCE" in row:  list_r.append("4")
        elif "KD-CRM" in row:  list_r.append("5")
        elif "KD-PROC" in row:  list_r.append("6")
        elif "KD-FACT" in row:  list_r.append("7")
        else: list_r.append(" ")
    df['PRODUCTO'] = list_r
    return df
 
Data_Proyectos =  add_product(Data_Proyectos)  
```
